### 4. Limpieza y Preprocesamiento de Datos

La limpieza de datos es un paso crucial en cualquier análisis de datos. Aquí, identificamos y manejamos valores nulos, corregimos tipos de datos incorrectos, eliminamos duplicados y filtramos datos relevantes. Este proceso asegura que los datos sean consistentes, precisos y listos para un análisis riguroso.

#### a. Identificación de los valores nulos

```python
# Revisión de valores nulos en cada dataset de nuevo
print("Valores nulos en Energy Dataset:")
print(energy_df.isnull().sum())

print("\nValores nulos en Weather Dataset:")
print(weather_df.isnull().sum())

print("\nValores nulos en Renewable Energy Dataset:")
print(consumption_df.isnull().sum())
```

#### b. Limpieza de Datos

##### b1. Energy Dataset

```python
# Paso 1: Eliminar columnas irrelevantes o con valores nulos masivos en Energy Dataset
# Mantener solo las columnas relevantes mencionadas
columns_to_keep_energy = [
    'time',
    'generation biomass',
    'generation fossil brown coal/lignite',
    'generation fossil gas',
    'generation nuclear',
    'generation solar',
    'generation wind onshore',
    'forecast solar day ahead',
    'forecast wind onshore day ahead',
    'total load forecast',
    'total load actual',
    'price day ahead',
    'price actual'
]
energy_df = energy_df[columns_to_keep_energy]

# Paso 2: Eliminar filas con valores nulos en Energy Dataset
# Como los valores nulos son menos del 1%, se pueden eliminar sin afectar mucho los datos.
energy_df = energy_df.dropna()
```

##### b2. Country and Weather Dataset

```python
# Paso 3: Limpiar Renewable Energy Dataset, seleccionando solo los cinco países más importantes
# Seleccionamos: China, United States, Germany, France, y Spain
countries_to_keep_renewable = ['Year', 'China', 'United States', 'Germany', 'France', 'Spain']
consumption_df = consumption_df[countries_to_keep_renewable]

# Paso 4: Eliminar filas con valores nulos en Renewable Energy Dataset
consumption_df = consumption_df.dropna()

# Paso 5: Limpieza de Weather Dataset
# Dado que no hay valores nulos en el Weather Dataset, no es necesario eliminar filas.
# Sin embargo, si en algún análisis futuro se identifican columnas irrelevantes, se podrían eliminar aquí.

# Guardar los datasets limpios
energy_df.to_csv('clean_energy_dataset.csv', index=False)
consumption_df.to_csv('clean_renewable_energy_dataset.csv', index=False)
weather_df.to_csv('clean_weather_dataset.csv', index=False)
```

#### c. Corrección de Tipos de Datos Incorrectos

En esta etapa, corregimos los tipos de datos incorrectos, como convertir las fechas en el formato adecuado y eliminar la zona horaria. Esto asegura que los datos se puedan utilizar de manera efectiva en análisis posteriores.

```python
# Conversión de tipos de datos
# Convertir la columna de fechas en el formato adecuado y eliminando la zona horaria
energy_df['time'] = pd.to_datetime(energy_df['time'], utc=True).dt.tz_localize(None)
weather_df['dt_iso'] = pd.to_datetime(weather_df['dt_iso'], utc=True).dt.tz_localize(None)
```

### 5. Análisis *Univariado*

El análisis univariado se enfoca en analizar la distribución de cada variable individual en los conjuntos de datos. Utilizamos gráficos como histogramas y gráficos de densidad para comprender la distribución, la dispersión y la tendencia central de cada variable. Este análisis proporciona información sobre el comportamiento de cada variable de manera independiente.

El análisis univariado implica explorar cada variable individualmente. Utilizaremos histogramas, gráficos de caja y gráficos de barras para visualizar la distribución de las variables numéricas y categóricas.

```python
import matplotlib.pyplot as plt

# Selección de los países a graficar
countries = ['Spain', 'China', 'United States', 'Germany', 'France']

# Convertir la columna Year en enteros
consumption_df['Year'] = consumption_df['Year'].astype(int)

# Graficar el consumo de energía de los países seleccionados
plt.figure(figsize=(14, 8))
for country in countries:
    plt.plot(consumption_df['Year'], consumption_df[country], marker='o', label=country)

# Configuración de la gráfica
plt.title('Consumo de Energía de Varios Países a lo Largo de los Años')
plt.xlabel('Año')
plt.ylabel('Consumo de Energía (TWh)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Analizamos la columna 'generation fossil gas' en energy_df
plt.figure(figsize=(10,6))
energy_df['generation fossil gas'].hist(bins=30, color='blue', edgecolor='black')
plt.title('Distribución de la Generación de Gas Fósil (kWh)')
plt.xlabel('Generación de Gas Fósil (kWh)')
plt.ylabel('Frecuencia')
plt.show()

# Analizamos la columna 'temp' en weather_df
plt.figure(figsize=(10,6))
weather_df['temp'].hist(bins=30, color='orange', edgecolor='black')
plt.title('Distribución de la Temperatura (K)')
plt.xlabel('Temperatura (K)')
plt.ylabel('Frecuencia')
plt.show()

# Consumo de energía de España a lo largo de los años en consumption_df
plt.figure(figsize=(10,6))
plt.plot(consumption_df['Year'], consumption_df['Spain'], marker='o', color='green')
plt.title('Consumo de Energía en España a lo Largo de los Años')
plt.xlabel('Año')
plt.ylabel('Consumo de Energía (España)')
plt.show()
```

### 6. Análisis Bivariado
Este paso explora las relaciones entre pares de variables para identificar correlaciones y patrones significativos. 

#### Ahora analizaremos la relación entre temperatura y generación de gas fósil, y luego crearemos matrices de correlación para cada dataset.

```python
import seaborn as sns

# Relación entre temperatura (weather_df) y generación de gas fósil (energy_df)
plt.figure(figsize=(10,6))
sns.scatterplot(x=weather_df['temp'], y=energy_df['generation fossil gas'])
plt.title('Relación entre Temperatura y Generación de Gas Fósil')
plt.xlabel('Temperatura (K)')
plt.ylabel('Generación de Gas Fósil (kWh)')
plt.show()

# Matriz de correlación para 'energy_df'
plt.figure(figsize=(10,8))
correlacion_energy = energy_df.corr()
sns.heatmap(correlacion_energy, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación - energy_df')
plt.show()
```

#### Gráfico de dispersión mostrando la correlación entre temperatura y generación solar

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarnos de que las columnas de tiempo estén en formato datetime
energy_df['time'] = pd.to_datetime(energy_df['time'])
weather_df['dt_iso'] = pd.to_datetime(weather_df['dt_iso'])

# Unir datasets por tiempo
merged_df = pd.merge(energy_df, weather_df, left_on='time', right_on='dt_iso', how='inner')

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=merged_df['temp'], y=merged_df['generation solar'])
plt.title('Correlación entre Temperatura y Generación Solar')
plt.xlabel('Temperatura (K)')
plt.ylabel('Generación Solar (kWh)')
plt.show()
```

### 7. Análisis Multivariado
El análisis multivariado implica el análisis de más de dos variables simultáneamente. Utilizamos técnicas como el Análisis de Componentes Principales (PCA) y clustering (agrupamiento) para reducir la dimensionalidad y encontrar patrones en datos complejos. Este análisis es útil para simplificar los datos y para entender cómo múltiples variables interactúan entre sí.

##### Aqui se puede ver claramente la correlación y como afecta una variable a la otra. Un buen ejemplo a mencionar es que la generación de energía solar y eólica se ve afectada por la temperatura y la humedad, haciendo que tengan una correlación negativa, o en otras palabras, inversamente proporcional.

```python
# Unir datasets por tiempo
merged_df = pd.merge(energy_df, weather_df, left_on='time', right_on='dt_iso', how='inner')
# Correlación entre clima y generación solar/eólica
sns.heatmap(merged_df[['generation solar', 'generation wind onshore', 'temp', 'humidity']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlación entre clima y generación solar/eólica')
plt.show()
```

#### Analisis de la evolución de la carga total de toda la energia y el precio de la energía en el tiempo, presentando adicionalmente la evolución del precio diario adelantado en el año 2015. Se puede observar en como tanto en la gráfica de varios años como en la del 2015 se sigue teniendo un techo de precion que no supera los 80 euros, y que la carga total de energía se mantiene en un rango de 0 a 40,000 MW.

```python
plt.figure(figsize=(12,6))
plt.plot(energy_df['time'], energy_df['total load actual'], label='Total Load Actual')
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.plot(energy_df['time'], energy_df['price day ahead'], label='Price Day Ahead')
plt.title('Evolución de la Carga Total y el Precio')
plt.legend()
plt.show()

# Filtrar los datos para el año 2015
energy_2015_df = energy_df[energy_df['time'].dt.year == 2015]

# Visualización de 'price day ahead' para el año 2015
plt.figure(figsize=(12, 6))
plt.plot(energy_2015_df['time'], energy_2015_df['price day ahead'], label='Price Day Ahead 2015')
plt.title('Evolución del Precio Diario Adelantado en 2015')
plt.xlabel('Tiempo')
plt.ylabel('Precio Day Ahead')
plt.legend()
plt.show()
```
